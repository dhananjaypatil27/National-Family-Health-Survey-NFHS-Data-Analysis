CHanging Data types (Categorical--> numeric)

```python
df_dtypes = df.dtypes.reset_index()
df_dtypes.columns = ['columns', 'dtypes']

df_dtypes.head()
```

```python
# Define categorical columns
cat_cols = ['District Names', 'State/UT']

# Clean and convert all other columns
for col in df.columns:
    if col not in cat_cols:
        df[col] = (
            df[col]
            .astype(str)                         # ensure string
            .str.replace(r'[^\d.]', '', regex=True)  # remove all non-numeric chars
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')  # convert to numeric

# Strip spaces from categorical columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.str.strip())

# Check result
print(df.dtypes.value_counts())
```

Null values

```python
# Null values in data
Null_values= df.isna().sum().sort_values(ascending=False)/len(df)*100
df_null_val= Null_values.reset_index()
df_null_val.columns=['columns','Null Values']
df_null_val.head()
```

```python
# Separate numeric & categorical
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include='object').columns

# Fill numeric with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical with mode
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Null values in data
Null_values= df.isna().sum().sort_values(ascending=False)/len(df)*100
df_null_val= Null_values.reset_index()
df_null_val.columns=['columns','Null Values']
df_null_val.head()
```

Outliers

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

outliers_summary = {}
for col in numeric_cols:
    outliers = detect_outliers(df, col)
    outliers_summary[col] = len(outliers)
```

```python
# Step 1: Convert dictionary to DataFrame
df_outliers = pd.DataFrame(
    list(outliers_summary.items()),
    columns=['columns', 'outlier no.']
)

# Step 2: Convert to numeric (VERY IMPORTANT)
df_outliers['outlier no.'] = pd.to_numeric(df_outliers['outlier no.'], errors='coerce')

# Step 3: Sort and ASSIGN BACK
df_outliers = df_outliers.sort_values(by='outlier no.', ascending=False)

# Step 4: Reset index (to see proper order)
df_outliers = df_outliers.reset_index(drop=True)

# Step 5: Display
df_outliers.head()
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_extreme = Q1 - 1.5 * IQR
    upper_extreme = Q3 + 1.5 * IQR

    # Cap extreme values
    df[col] = np.where(df[col] < lower_extreme, lower_extreme, df[col])
    df[col] = np.where(df[col] > upper_extreme, upper_extreme, df[col])
```
